## Create Table and DB in Python for SQL   

```python
import sqlite3
```

```python
connection = sqlite3.connect('LeMastertTech.db') # establishes connection 
cursor = connection.cursor() #builtin of SQL package. lets you see and recieve commands back and forth
```

## Each Entry will have - name, views, date (YYYY/MM/DD)

SQL Command inside of triple quotes (for long blocks of text in python) - MUST be written in SQL format

```python
#cursor.execute("DROP TABLE SAMPLE") THIS IS ONLY USED IF THERE ALREADY IS A TABLE CALLED SAMPLE.
sql_command = """ CREATE TABLE sample(
    vid_name VARCHAR(30),
    views INTEGER,
    published DATE
)"""

cursor.execute(sql_command) # executes the command to create the table
#connection.close()

```

```python
videos = [['pacman', 17704, '2022-10-25'],
          ['read from text', 49677, '2021-12-29'],
          ['controls top 5', 15076, '2022-03-31'],
          ['python excel', 8037, '2022-05-19']]
```

INSERT INTO table

```python
for i in range(len(videos)):
    name = videos[i][0]
    views = videos[i][1]
    date = videos[i][2]
    cursor.execute("INSERT INTO sample VALUES (?, ?, ?)", (name, views, date))
    
connection.commit() # commits the changes to the database
```

VIEW (SELECT) Table

```python
cursor.execute("SELECT * FROM sample")
results = cursor.fetchall() # fetches all the results from the query
for i in results:
    print(i)
```

```python
import pandas as pd
df = pd.read_sql_query("SELECT * FROM sample", connection)
print(df.columns.tolist())
print(df)
```
