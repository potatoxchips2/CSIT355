## Each Entry will have - name, views, date (YYYY/MM/DD)

SQL Command inside of triple quotes (for long blocks of text in python) - MUST be written in SQL format

```python
#cursor.execute("DROP TABLE SAMPLE") THIS IS ONLY USED IF THERE ALREADY IS A TABLE CALLED SAMPLE.
sql_command = """ CREATE TABLE sample(
    vid_name VARCHAR(30),
    views INTEGER,
    published DATE
)"""

cursor.execute(sql_command) # executes the command to create the table
#connection.close()

```

```python
videos = [['pacman', 17704, '2022-10-25'],
          ['read from text', 49677, '2021-12-29'],
          ['controls top 5', 15076, '2022-03-31'],
          ['python excel', 8037, '2022-05-19']]
```

INSERT INTO table

```python
for i in range(len(videos)):
    name = videos[i][0]
    views = videos[i][1]
    date = videos[i][2]
    cursor.execute("INSERT INTO sample VALUES (?, ?, ?)", (name, views, date))
    
connection.commit() # commits the changes to the database
```

###Displaying the content of Database (.db)

```python
cursor.execute(""" SELECT name, type FROM sqlite_master WHERE type IN ('table','view') AND name NOT LIKE 'sqlite_%' ORDER BY type, name; """)
objects = cursor.fetchall()
print("Objects:", objects)
Objects: [('Department', 'table'), ('Employee', 'table'), ('Reserves','table'), ('Sailors', 'table'), ('sample', 'table')]
for name, typ in objects:
    if typ != "table":
        continue
    print(f"\n=== TABLE: {name} ===")
    # Columns
    cursor.execute(f"PRAGMA table_info({name});")
    cols = [r[1] for r in cursor.fetchall()]
    print("Columns:", cols)
    
    # Row count
    cursor.execute(f"SELECT COUNT(*) FROM {name};")
    print("Rows:", cursor.fetchone()[0])
    # First 5 rows
    cursor.execute(f"SELECT * FROM {name} LIMIT 5;")
    for row in cursor.fetchall():
        print(row)
```

###Sqlite doesn't support ANY or ALL - We can write these queries with Join. Will cover in later chapters

```python
import sqlite3
# Use ':memory:' for an in-memory DB, or replace with 'boats.db' to save to a file
conn = sqlite3.connect(':memory:')
cur = conn.cursor()
cur.executescript(""" 
-- Drop if they exist (for repeatable runs) 
DROP TABLE IF EXISTS Reserves;
DROP TABLE IF EXISTS Boats;
DROP TABLE IF EXISTS Sailors;
-- Create tables
CREATE TABLE Sailors (
    sid INTEGER PRIMARY KEY,
    sname TEXT NOT NULL
);
CREATE TABLE Boats (
    bid INTEGER PRIMARY KEY,
    color TEXT
);
CREATE TABLE Reserves (
    sid INTEGER,
    bid INTEGER, 
    FOREIGN KEY (sid) REFERENCES Sailors(sid),
    FOREIGN KEY (bid) REFERENCES Boats(bid)
);
-- Insert sample data
INSERT INTO Sailors (sid, sname) VALUES
    (1, 'Alice'),
    (2, 'Bob'),
    (3, 'Cara'),
    (4, 'Dan');
INSERT INTO Boats (bid, color) VALUES
    (101, 'red'),
    (102, 'green'),
    (103, 'red');
INSERT INTO Reserves (sid, bid) VALUES
    (1, 101),
    (1, 102),
    (1, 103),
    (2, 101),
    (2, 103),
    (3, 101),
    (3, 102),
    (3, 103),
    (4, 102);
""")
conn.commit()
# (Optional) quick checks
print("Sailors:", cur.execute("SELECT * FROM Sailors").fetchall())
print("Boats:", cur.execute("SELECT * FROM Boats").fetchall())
print("Reserves:", cur.execute("SELECT * FROM Reserves").fetchall())
cur.close()
conn.close()
```
