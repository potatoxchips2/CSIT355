FIND AVG AGE OF EMP FOR EACH DEPT WHERE EVERY EMP IS 30 OR YOUNGER. LIST DEPT ID AS WELL

```python
%%sql 
SELECT D.did, AVG(E.AGE)
FROM Employee E, Works W, Department D
WHERE E.eid = W.eid AND W.did = D.did
GROUP BY D.did
HAVING 0 = (
SELECT COUNT(*)
FROM Employee E1, Works W1
WHERE E1.eid = W1.eid AND W1.did = D.did AND E1.Age > 30
)
```

FIND EACH MANAGER BY EID THE AVG EMP SALARY

```python
%%sql
SELECT D.managerid, D.did, AVG(E.salary)
FROM Employee E, Works W, Department D
WHERE E.eid = W.eid AND W.did = D.did
GROUP BY D.managerid
```
